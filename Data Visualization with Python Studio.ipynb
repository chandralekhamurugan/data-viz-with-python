```python
# Number of books published per year.

# Use your Python skills to create a visualization here.
# Grouped bar chart


df = pd.read_csv("books.csv", on_bad_lines='skip')
df['year'] = df.sum(axis = 1, numeric_only=True)


years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years



df_group = df.groupby('year')[['authors', 'isbn']].count()

# Grouped bar chart
df_group.plot.bar(y='authors', edgecolor='white')
plt.title('Number of books published per year', color='black')
plt.xticks(rotation=0, color='black')
plt.yticks(color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Number of Books', color='black')
plt.legend().remove()
plt.show()
```

```python
df_group = df.groupby(['publisher'])['title'].count().sort_values(ascending=False)
df_group
```

```python
# Number of books published by each publisher

# Use your Python skills to create a visualization here.
# Vertical bar chart

style.use('ggplot')

#df_group = pd.DataFrame(df.loc[:,years].sum()).T
# df_group.rename(columns = {0:'total'}, inplace = True)


df_group.plot(kind = 'bar', legend = False)
plt.title('Number of books published by each publisher',color = 'black')
#plt.xticks(color = 'black')
#plt.yticks(color = 'black')
plt.legend(loc='upper right',fontsize =12)
plt.xlabel('Publisher',color = 'black')
plt.ylabel('title',color = 'black')
plt.savefig('bar_vertical.png')
plt.show()
```

```python
# Use your Python skills to create a visualization here.
# cyberpunk simple area chart
import mplcyberpunk

style.use('cyberpunk')

plt.fill_between(df_group.index, df_group['title'], color = 'greenyellow', alpha = 0.1)
plt.plot(df_group.index, df_group['title'], color = 'palegreen', alpha = 1)
mplcyberpunk.add_glow_effects()
plt.title('Number of books published by each publisher', fontsize = 20,fontweight = 'bold', color = 'white')
plt.yticks(fontsize = 14, color = 'white',fontweight = 'bold')
plt.xticks(fontsize = 14, rotation = 90, color = 'white',fontweight = 'bold')
plt.xlabel('Publisher', fontsize = 16, color = 'white',fontweight = 'bold')
plt.ylabel('Title', fontsize = 16, color = 'white',fontweight = 'bold')
plt.savefig('cyber_area_simple.png')

plt.show()
```

```python
# Text ratings count versus reviews count.

# Use your Python skills to create a visualization here.

df = pd.read_csv("books.csv", on_bad_lines='skip')

# Scatter plot
style.use('fivethirtyeight')
plt.scatter(df['text_reviews_count'], df['ratings_count'], alpha=0.5)
plt.title('Text Reviews Count vs Ratings Count', color='black')
plt.xlabel('Text Reviews Count', color='black')
plt.ylabel('Ratings Count', color='black')
plt.xticks(color='black')
plt.yticks(color='black')
plt.show()
```
